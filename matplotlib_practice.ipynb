```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#5.1.4 skill drill
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis,label = "Boston", color ='green', marker = 'd', linewidth = (2))
ax.set_title("PyBer Fare Month")
ax.set_xlabel('Month')
ax.set_ylabel('Amount($)')
ax.set_ylim(0,50)
ax.grid()
ax.legend()
```

```python
#using loops to adjust the size
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s= y_axis_larger)    
```

```python
# object oriented interface method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# skill drill 5.1.7
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, label="Boston", c='skyblue', alpha=0.8, edgecolors= "k", s=[i*5 for i in y_axis])
plt.gca().invert_yaxis()
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0,50)
```

```python
import numpy as np
```

```python
#changing major tickers using function methos
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step= 5.0))
plt.gca().invert_yaxis()
```

```python
#changing major tickers using object oriented method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,50,5))
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
